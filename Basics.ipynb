```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

# Generate line charts to compare across all degree categories

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig1 = plt.figure(figsize=(18, 18))
# generate first coulmn of line charts , STEM . 
for sp in range (0,18,3): 
    sp_cats = int(sp/3)
    ax = fig1.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp_cats]], c=cb_dark_blue, label="Women", 
                                           linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp_cats]], c=cb_orange, label="Men", 
                                            linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2010)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(stem_cats[sp_cats])
    ax.tick_params(bottom="off", top="off", left="off", 
                   right="off", labelbottom='off')   # disable the x-axis labels
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)    # add horizontal line
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 15:
        ax.text(2005, 87, 'Men')
        ax.text(2001, 10, 'Women') 
        ax.tick_params(labelbottom='on') # enable the x-axis labels for the bottommost line charts
# generate second coulmn of line charts , liberal arts . 
for sp in range (1,16,3):
    sp_cats = int((sp-1)/3)
    ax = fig1.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp_cats]], c=cb_dark_blue, label="Women", 
                                           linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp_cats]], c=cb_orange, label="Men", 
                                            linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2010)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(lib_arts_cats[sp_cats])
    ax.tick_params(bottom="off", top="off", left="off", 
                   right="off", labelbottom='off')  # disable the x-axis labels
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)   # add horizontal line
    if sp == 1:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')          
    elif sp == 13:
        ax.tick_params(labelbottom='on')
# generate third coulmn of line charts , other categroies .                                                      
for sp in range (2,20,3):
    sp_cats = int((sp-2)/3)
    ax = fig1.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp_cats]], c=cb_dark_blue, label="Women", 
                                           linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp_cats]], c=cb_orange, label="Men", 
                                            linewidth=3) 
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2010)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(other_cats[sp_cats])
    ax.tick_params(bottom="off", top="off", left="off", 
                   right="off", labelbottom='off')   # disable the x-axis labels
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)  # add horizontal line
    if sp == 2:
        ax.text(2005, 90, 'Men')
        ax.text(2002, 5, 'Women')
    elif sp == 17:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 35, 'Women')  
        ax.tick_params(labelbottom='on')
plt.savefig('gab_gender_degrees.png')                
plt.show()
```
